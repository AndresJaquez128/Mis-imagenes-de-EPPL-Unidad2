#### Redaccion y contexto del problema
En el estudio de la geometría, el círculo y la esfera son figuras fundamentales
que encontramos constantemente en nuestra vida diaria. Desde las ruedas de un
automóvil hasta los planetas en el espacio, estas formas geométricas están
presentes en múltiples aplicaciones.

El círculo es una figura plana bidimensional donde todos los puntos equidistan
de un punto central llamado centro. La distancia desde el centro hasta cualquier
punto del círculo se denomina radio.

La esfera es la versión tridimensional del círculo, donde todos los puntos en
su superficie están a la misma distancia del centro.

Para realizar cálculos con estas figuras, necesitamos conocer fórmulas
matemáticas específicas que nos permiten determinar sus propiedades principales..

![Circulo](https://raw.githubusercontent.com/AndresJaquez128/Mis-imagenes-de-EPPL-Unidad2/refs/heads/main/circulo.PNG)

1️⃣  ÁREA DEL CÍRCULO:
    ──────────────────
    A = π × r²
    
    Donde:
    • A = Área del círculo
    • π = Pi (3.141592653589793...)
    • r = radio del círculo

2️⃣  PERÍMETRO DEL CÍRCULO (Circunferencia):
    ───────────────────────────────────────
    P = 2 × π × r
    
    Donde:
    • P = Perímetro o circunferencia
    • π = Pi
    • r = radio del círculo

3️⃣  VOLUMEN DE LA ESFERA:
    ─────────────────────
    V = (4/3) × π × r³
    
    Donde:
    • V = Volumen de la esfera
    • π = Pi
    • r = radio de la esfera

```python
# Importamos la librería math para usar el valor de pi
import math

# Constante global
PI = math.pi

def areacirculo():
    """
    Función para calcular el área de un círculo
    Fórmula: A = π × r²
    """
    print("\n" + "🔵" * 40)
    print("🔵           CÁLCULO DEL ÁREA DEL CÍRCULO           🔵")
    print("🔵" * 40)
    print("\n📐 Fórmula: A = π × r²")
    print("-" * 40)

    # Solicitar el radio
    radio = float(input("\n📏 Ingresa el radio del círculo: "))

    # Calcular el área
    area = PI * (radio ** 2)

    # Mostrar resultados
    print("\n" + "✨" * 40)
    print("📊 RESULTADOS:")
    print(f"   • Radio ingresado: {radio}")
    print(f"   • Fórmula: A = π × ({radio})²")
    print(f"   • 🎯 Área del círculo: {area:.4f} unidades²")
    print("✨" * 40)

def perimetrocirculo():
    """
    Función para calcular el perímetro de un círculo
    Fórmula: P = 2 × π × r
    """
    print("\n" + "🔴" * 40)
    print("🔴        CÁLCULO DEL PERÍMETRO DEL CÍRCULO         🔴")
    print("🔴" * 40)
    print("\n📐 Fórmula: P = 2 × π × r")
    print("-" * 40)

    # Solicitar el radio
    radio = float(input("\n📏 Ingresa el radio del círculo: "))

    # Calcular el perímetro
    perimetro = 2 * PI * radio

    # Mostrar resultados
    print("\n" + "✨" * 40)
    print("📊 RESULTADOS:")
    print(f"   • Radio ingresado: {radio}")
    print(f"   • Fórmula: P = 2 × π × {radio}")
    print(f"   • 🎯 Perímetro del círculo: {perimetro:.4f} unidades")
    print("✨" * 40)

def volumencirculo():
    """
    Función para calcular el volumen de una esfera
    Fórmula: V = (4/3) × π × r³
    """
    print("\n" + "🔷" * 40)
    print("🔷          CÁLCULO DEL VOLUMEN DE LA ESFERA          🔷")
    print("🔷" * 40)
    print("\n📐 Fórmula: V = (4/3) × π × r³")
    print("-" * 40)

    # Solicitar el radio
    radio = float(input("\n📏 Ingresa el radio de la esfera: "))

    # Calcular el volumen
    volumen = (4/3) * PI * (radio ** 3)

    # Mostrar resultados
    print("\n" + "✨" * 40)
    print("📊 RESULTADOS:")
    print(f"   • Radio ingresado: {radio}")
    print(f"   • Fórmula: V = (4/3) × π × ({radio})³")
    print(f"   • 🎯 Volumen de la esfera: {volumen:.4f} unidades³")
    print("✨" * 40)

def menu():
    """
    Función principal que muestra el menú y controla el programa
    """
    opcion = 0

    while opcion != 4:
        # Mostrar menú
        print("\n" + "🌟" * 40)
        print("🌟           MENÚ PRINCIPAL - GEOMETRÍA           🌟")
        print("🌟" * 40)
        print("\n         📐 CÁLCULOS GEOMÉTRICOS 📐")
        print("\n         1️⃣  Calcular área del círculo")
        print("         2️⃣  Calcular perímetro del círculo")
        print("         3️⃣  Calcular volumen de la esfera")
        print("         4️⃣  Salir del programa")
        print("\n" + "🌟" * 40)

        # Leer opción
        opcion = int(input("\n💡 Selecciona una opción (1-4): "))

        # Estructura if-elif-else para manejar las opciones
        if opcion == 1:
            areacirculo()
        elif opcion == 2:
            perimetrocirculo()
        elif opcion == 3:
            volumencirculo()
        elif opcion == 4:
            print("\n" + "🎉" * 40)
            print("🎉                                              🎉")
            print("🎉         ¡GRACIAS POR USAR EL PROGRAMA!      🎉")
            print("🎉              👋 ¡Hasta pronto! 👋            🎉")
            print("🎉                                              🎉")
            print("🎉" * 40)
        else:
            print("\n" + "⚠️" * 40)
            print("⚠️                                              ⚠️")
            print("⚠️            ⚠️ OPCIÓN NO VÁLIDA ⚠️             ⚠️")
            print("⚠️     Por favor, selecciona una opción del 1 al 4     ⚠️")
            print("⚠️                                              ⚠️")
            print("⚠️" * 40)

# ==================== PROGRAMA PRINCIPAL ====================
if __name__ == "__main__":
    # Mostrar la redacción y fórmulas iniciales
    print(__doc__)

    # Pausa para leer
    input("💾 Presiona ENTER para continuar al menú principal...")

    # Llamar a la función del menú
    menu()
```
